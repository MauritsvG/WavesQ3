<h3 style="color:#00BFFF;">Offset removal  </h3>
<hr style="height:1.5px;border:none;color:#333;" />


Remove the vertical offset in the timeseries, i.e. transform the time series so that the elevation and velocities are zeros when the water is at rest. First, visually inspect the data to see for what time interval this is still the case. Then determine the offset, for example via the mean elevation or velocity over this time interval. <h3 style="color:#00BFFF;">Data calibration </h3>
<hr style="height:1.5px;border:none;color:#333;" />

The next step is to convert from volts to flow velocities for EMF and from volts to meters for WG. Every EMF follows the same conversion formula $u=0.1023x$, where $x$ is the data read. Do not worry yet about the initial offset (i.e.,flow velocities not being 0 at the beginning of the run), that will be removed in the next step.

Different calibration parameters are used for converting the wave gauge data to metres. Use the following formula $\eta=coefficient * x$, where $x$ is the data in Volts and $\eta$ is the elevation defined with respect to the still water level.

<h3 style="color:#00BFFF;"> Orientation of the velocimeter </h3>
<hr style="height:1.5px;border:none;color:#333;" />

If you are working with the velocity timeseries, make sure that the instruments have been orientated properly (x-component of the velocity positive towards the shore). One quick way to do so is to check that the horizontal velocity and elevation at the start of the run (before any reflection takes place) are in phase. For each measurement station (x-location), plot the first seconds of the data for both the WG and EMF on the same figure and examine the phasing of the oscillations. The scales of the two signals should be different, but the first waves should be in phase. If they are 180 degrees out of phase  (e.g., elevation reaches its maximum when velocity reaches a minimum), this means that the x-axis of the velocimeter is oriented offshore; in that case, simply multiply the output by -1 and proceed. 

<h3 style="color:#00BFFF;"> Stationarity  </h3>
<hr style="height:1.5px;border:none;color:#333;" />

The wave paddle starts generating waves at the offshore boundary over an initially flat water surface. It will take some time before stationary conditions are reached (e.g., the waves need to propagate over the full domain and back in case of reflection, and setup needs time to build up). Do not forget to remove this "spinup time" before calculating wave statistics or the wave-induced setup. 
<br>
To make a first estimate of the the spinup time, estimate the time needed for the waves to reach the beach, and then multiply it by 2 or 3. This is the minimum period that you should remove from all your timeseries. Also remove the last part of the run, once the wave paddle has stopped.
